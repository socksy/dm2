```python
# Remove the columns we do not need to use for this project

# removing columns from a data set via column names
def columnRemoval(table,columnNames):
    for i in range(0, len(columnNames)):
        table = table.drop(columnNames[i], 1)
    return table
 
removal = ['VehYear','WheelTypeID','TopThreeAmericanName','Nationality','VNZIP1','BYRNO', 'PRIMEUNIT', 'AUCGUART']
auto = columnRemoval(auto, removal)
autotest = columnRemoval(autotest,removal)
```

```python
# Combine ALL the average prices

# round to nearest integer
def roundToNearest(num, nearest):
    return round(num/(nearest*1.0))*(nearest*1.0)

def setAvgPrices(table):
    avgAcquisitionPrices = []
    avgCurrentPrices = []
    
    # for each row
    for i in range(0, len(table.RefId)):
        
        # get the average aquisition prices
        totalAvgPrice = 0
        totalAvgPrice = totalAvgPrice + table.MMRAcquisitionAuctionAveragePrice[i]
        totalAvgPrice = totalAvgPrice + table.MMRAcquisitionAuctionCleanPrice[i]
        totalAvgPrice = totalAvgPrice + table.MMRAcquisitionRetailAveragePrice[i]
        totalAvgPrice = totalAvgPrice + table.MMRAcquisitonRetailCleanPrice[i]
        avgAcquisitionPrices.append(roundToNearest(totalAvgPrice/4.0, 500))
        
        # get the average current prices
        totalAvgPrice = 0
        totalAvgPrice = totalAvgPrice + table.MMRCurrentAuctionAveragePrice[i]
        totalAvgPrice = totalAvgPrice + table.MMRCurrentAuctionCleanPrice[i]
        totalAvgPrice = totalAvgPrice + table.MMRCurrentRetailAveragePrice[i]
        totalAvgPrice = totalAvgPrice + table.MMRCurrentRetailCleanPrice[i] 
        avgCurrentPrices.append(roundToNearest(totalAvgPrice/4.0, 500))     
     
    table['AverageAcquisitionPrice'] = avgAcquisitionPrices
    table['AverageCurrentPrice'] = avgCurrentPrices
    
    return table

auto = setAvgPrices(auto)
autotest = setAvgPrices(autotest)
```

```python
# Calculate the difference between the current and aquisition price
def priceDiffs(table):
    diffs = []
    for i in range(0, len(table.RefId)):
        # get the difference
        diff = table.AverageCurrentPrice[i] - table.AverageAcquisitionPrice[i]

        # get the % difference from the aquisition price
        av = round((100.0*diff)/table.AverageAcquisitionPrice[i])
        
        # if the average is 'nan'
        if math.isnan(float(av)) or math.isinf(float(av)) or av > 1000:
            av = 0.0
            
        diffs.append(av)
       
    return diffs

auto['PriceDiff'] = priceDiffs(auto)
autotest['PriceDiff'] = priceDiffs(autotest)
```

```python
# Add a column for depreciation % per year on average
# depreciation value = (((price sold for - original value)/age in years)/original value)*100.0


def addDepreciationValues(table):
    # depreciation value list
    depVals = []

    # for each row
    for i in range(0, len(table.RefId)):
        # get original value
        origVal = table.AverageAcquisitionPrice[i]

        # get price sold for
        soldFor = table.VehBCost[i]     

        # get age in years
        ageYears = table.VehicleAge[i]

        # get depreciation value
        depVal = (((soldFor - origVal)/ageYears)/origVal)*100.0

        # round the value to the nearest 5%
        depVal = roundToNearest(depVal, 1)

        # if the value is over 100% or infinity
        if depVal > 100 or math.isinf(depVal):
            depVal = 999.0

        # add the value to the list
        depVals.append(depVal)
        
    # add the column to the table
    table['DepreciationValue'] = depVals
    
    return table

auto = addDepreciationValues(auto)
autotest = addDepreciationValues(autotest)
```

```python
# Miles per year column
def addMilesPerYear(table):
    # miles list
    milesList = []

    # for each row
    for i in range(0, len(table.RefId)):
        # get original value
        origMiles = table.VehOdo[i]

        # get age in years
        ageYears = table.VehicleAge[i]

        # get depreciation value
        miles =  origMiles/ageYears

        # round the value to the nearest 1000 miles
        miles = roundToNearest(miles, 1000)

        # if the value is over 100% or infinity
        if math.isnan(miles) or math.isinf(miles):
            miles = -1

        # add the value to the list
        milesList.append(miles)
        
    # add the column to the table
    table['MilesPerYear'] = str(milesList) + "Miles/Year"
    
    print min(milesList)
    print max(milesList)
    return table

auto = addMilesPerYear(auto)
autotest = addMilesPerYear(autotest)
```

```python
# Convert PURCHDATE to a purchase season

# Define function that gets the season from a the purchase date
# month 12,1,2 is winter, 3,4,5 is sping, 6,7,8 is summer and then 9,10,11 is winter
def getSeason(string):
    month = string[0:2]
    if month in ['12','1/','2/']:
        return 'Winter'
    elif month in ['3/','4/','5/']:
        return 'Spring'
    elif month in ['6/','7/','8/']:
        return 'Summer'
    elif month in ['9/','10','11']:
        return 'Autumn'
    
    return 'Unknown'
    
# apply season function to Purch Date Date on auto and autotest
auto['PurchDate'] = auto.PurchDate.apply(getSeason)
autotest['PurchDate'] = autotest.PurchDate.apply(getSeason)
```

```python
# Split Model field into Car Model name, Engine Size (V cylinders), Wheel Drive, Straight Cylinders

# remove a sub string from a string and the space it leaves behind
def removeSubStringAndSpace(string, substring):
    string = string.replace(substring,'')
    string = string.replace('  ',' ')
    string = string.lstrip(' ')
    string = string.rstrip(' ')
    return string

# remove engine size (V6, V8, etc.)
# return the engine size (if it exists in the string) and remove it from the string
def getEngineSize(string):
    if 'V6' in string:
        string = removeSubStringAndSpace(string, 'V6')
        return (6, string)    
    elif 'V8' in string:
        string = removeSubStringAndSpace(string, 'V8')
        return (8, string)
   
    else: 
        return (-1, string)

# remove wheel drive (2WD, 4WD, FWD)
# return the wheel drive (if it exists in the string) and remove it from the string
def getWheelDrive(string):
    if '2WD' in string:
        string = removeSubStringAndSpace(string, '2WD')
        return ('2WD', string)    
    elif '4WD' in string:
        string = removeSubStringAndSpace(string, '4WD')
        return ('4WD', string)
    
    elif 'FWD' in string:
        string = removeSubStringAndSpace(string, 'FWD')
        return ('FWD', string)
    
    else: 
        return ('OTHER', string)

# remove cylinders (4C or 6C)    
# return the number of cylinders (if it exists in the string) and remove it from the string
def getNumberCylinders(string):
    if '4C' in string:
        string = removeSubStringAndSpace(string, '4C')
        return (4, string)    
    elif '6C' in string:
        string = removeSubStringAndSpace(string, '6C')
        return (6, string)
    
    else: 
        return (-1, string)

    
# return the number of litres (if it exists in the string), 'None' otherwise
def getLitresSubstring(string):
    # return any substring that matches a digit followed by a . followed by nothing or digits or an L
    m = re.search('\d\.\d*L*', string)
    if m:
        return m.group(0)
    
    return 'None'

# remove the litres from a string
def removeLitresFromString(string):
    # get if they are litres are in the string
    litres = getLitresSubstring(string)
    if 'None' in litres:
        return string
    
    # else remove the litres from the string
    return removeSubStringAndSpace(string, litres)  
   
# split the Model column in a table
def splitModelColumn(table):
    # create 3 new arrays for the values to add
    engineSizes = []
    wheelDrives  = []
    cylinders = []
    models = []
    
    # for each row in the models
    for i in range(0, len(table.Model)):
        # get the next model name
        s = table.Model[i]
        
        # strip the engine size
        s = getEngineSize(s)
        engineSizes.append(s[0])
       
        # strip wheel drive
        s = getWheelDrive(s[1])
        wheelDrives.append(s[0])
        
        # strip number of cylinders
        s = getNumberCylinders(s[1])
        cylinders.append(s[0])
        
        # remove any litres
        s = removeLitresFromString(s[1])
        
        # append the reduced model name to the models name
        models.append(s)
        
    # set the new model names
    table['Model'] = models
    
    # add the new columns to the table
    table['EngineSize'] = engineSizes
    table['WheelDrives'] = wheelDrives
    table['Cylinders'] = cylinders
    
    # return the new table with the split Model column
    return table

# apply the model column change to both the auto and autotest tables
auto = splitModelColumn(auto)
autotest = splitModelColumn(autotest)
```

```python
# Split SUB-MODEL into Number of doors and Litres

# return the number of doors 
def getNumberDoors(string):
    if '2D' in string or '3D' in string:
        return 2
    elif '4D' in string or '5D' in string:
        return 4
    
    else: 
        return -1
    
# convert a litres string to a float
def litresToFloat(litresString):
    if litresString.startswith('None'):
        return -1.0
    
    # if it ends with an L or l
    if litresString.endswith('L') or litresString.endswith('l'):
        litresString = litresString[:-1]
        return float(litresString)
    
    return float(litresString)
    
# split the Sub-Model column in a table
def splitSubModelColumn(table):
    # create new arrays for the values to add
    numberDoors = []
    litres  = []
    
    # for each row in the sub-models
    for i in range(0, len(table.SubModel)):
        # get the next sub model name
        submodel = table.SubModel[i]
        
        # if its a float, don't bother trying to cast it
        if isinstance(submodel, float):
            s = ""
        else:
            s = submodel
        
        # get the number doors and append to doors list
        doors = getNumberDoors(s)
        numberDoors.append(int(doors))
       
        # get the number of litres and append to litres list
        lits = getLitresSubstring(s)     
        
        # convert the litres to a float
        lits = litresToFloat(lits)
        
        litres.append(lits)
    
    # add the new columns to the table
    table['NumDoors'] = numberDoors
    table['Litres'] = litres
    
    # remove the Sub-Model column
    table = table.drop('SubModel', 1)
    
    # return the new table with the split Sub-Model column
    return table

# split the sub-model columns in the auto and autotest tables
auto = splitSubModelColumn(auto)
autotest = splitSubModelColumn(autotest)   
```

```python
# Convert WheelType strings to integers

# function to convert all values in a column to upper case
def toUpper(column):
    # for each value in the column
    for i in range(0, len(column)):
        
        # if the object is a string, convert to uppercase
        if isinstance(column[i], str):
            column[i] = column[i].upper()
    
    return column
        
# function to get unique values from a column
def uniqueValues(column):
    return set(column)

# return if a value is nan (float)
def isNan(thing):
     return isinstance(thing, float) and math.isnan(thing)

# replace punctuation in a list of strings
def replacePunc(stringList):
    regex = re.compile('[%s]' % re.escape(string.punctuation))
    for i in range(0, len(stringList)):
        stringList[i] = regex.sub(' ', stringList[i])
        
    return stringList
        
# replace a string by the longest word in it
def replaceWithLongest(stringList):
    for i in range(0, len(stringList)):
        substrings = stringList[i].split()
        longest = max(substrings, key=len)
        stringList[i] = longest
        
    return stringList

# remove punctuation, replace with longest word and reduce to a set
def reduceColumn(col):
    col = replacePunc(col)
    col = replaceWithLongest(col)
    return col
    
        
###############
# GENERIC function to convert a set of string values for a column into integers ##
###############

# convert to dummies and remove old column
def dummies(data, column_name):
    # get the column from the table
    column = data[column_name]
    
    new_variable = get_dummies(column)
    
    data = data.drop(column_name, axis=1)
    data = pd.concat([data, new_variable],axis=1)
   
    return data

# convert Purchase date for both auto and autotest
# auto = dummies(auto, 'PurchDate')
# autotest = dummies(autotest, 'PurchDate')

# convert auction for both auto and autotest
# auto = dummies(auto, 'Auction')
# autotest = dummies(autotest, 'Auction')
    
# convert Make for both auto and autotest
# auto = dummies(auto, 'Make')
# autotest = dummies(autotest, 'Make')

# reduce the model name column values
# auto['Model'] = reduceColumn(auto['Model'])
# autotest['Model'] = reduceColumn(autotest['Model'])

# convert model for both auto and autotest
# auto = dummies(auto, 'Model')
# autotest = dummies(autotest, 'Model')

# convert the tirm for both auto and autotest
auto = dummies(auto, 'Trim')
autotest = dummies(autotest, 'Trim')

# convert the colour for both auto and autotest
auto = dummies(auto, 'Color')
autotest = dummies(autotest, 'Color')

# convert the transmission columns to upper case
# AND convert the transmission for both auto and autotest
auto['Transmission'] = toUpper(auto.Transmission)
autotest['Transmission'] = toUpper(autotest.Transmission)
auto = dummies(auto, 'Transmission')
autotest = dummies(autotest, 'Transmission')

# convert the wheel type for both auto and autotest
auto = dummies(auto, 'WheelType')
autotest = dummies(autotest, 'WheelType')

# convert the number of doors for both auto and autotest
# auto = dummies(auto, 'NumDoors')
# autotest = dummies(autotest, 'NumDoors')

# convert the engine size for both auto and autotest
# auto = dummies(auto, 'EngineSize')
# autotest = dummies(autotest, 'EngineSize')

# convert the Wheel Drives for both auto and autotest
# auto = dummies(auto, 'WheelDrives')
# autotest = dummies(autotest, 'WheelDrives')

# convert the Cylinders for both auto and autotest
# auto = dummies(auto, 'Cylinders')
# autotest = dummies(autotest, 'Cylinders')

# convert the Litres for both auto and autotest
auto = dummies(auto, 'Litres')
autotest = dummies(autotest, 'Litres')

# convert the size for both auto and autotest
auto = dummies(auto, 'Size')
autotest = dummies(autotest, 'Size')

# convert state for both auto and autotest
# auto = dummies(auto, 'VNST')
# autotest = dummies(autotest, 'VNST')

# convert MilesPerYear for both auto and autotest
auto = dummies(auto, 'MilesPerYear')
autotest = dummies(autotest, 'MilesPerYear')
```
