```python
# Convert PURCHDATE to a purchase season

# Define function that gets the season from a the purchase date
# month 12,1,2 is winter, 3,4,5 is sping, 6,7,8 is summer and then 9,10,11 is winter
def getSeason(string):
    month = string[0:2]
    if month in ['12','1/','2/']:
        return 'Winter'
    elif month in ['3/','4/','5/']:
        return 'Spring'
    elif month in ['6/','7/','8/']:
        return 'Summer'
    elif month in ['9/','10','11']:
        return 'Autumn'
    
    return 'Unknown'
    
# apply season function to Purch Date Date on auto and autotest
auto['PurchDate'] = auto.PurchDate.apply(getSeason)
autotest['PurchDate'] = autotest.PurchDate.apply(getSeason)
```

```python
# Split Model field into Car Model name, Engine Size (V cylinders), Wheel Drive, Straight Cylinders

# remove a sub string from a string and the space it leaves behind
def removeSubStringAndSpace(string, substring):
    string = string.replace(substring,'')
    string = string.replace('  ',' ')
    string = string.lstrip(' ')
    string = string.rstrip(' ')
    return string

# remove engine size (V6, V8, etc.)
# return the engine size (if it exists in the string) and remove it from the string
def getEngineSize(string):
    if 'V6' in string:
        string = removeSubStringAndSpace(string, 'V6')
        return ('V6', string)    
    elif 'V8' in string:
        string = removeSubStringAndSpace(string, 'V8')
        return ('V8', string)
   
    else: 
        return ('OTHER', string)

# remove wheel drive (2WD, 4WD, FWD)
# return the wheel drive (if it exists in the string) and remove it from the string
def getWheelDrive(string):
    if '2WD' in string:
        string = removeSubStringAndSpace(string, '2WD')
        return ('2WD', string)    
    elif '4WD' in string:
        string = removeSubStringAndSpace(string, '4WD')
        return ('4WD', string)
    
    elif 'FWD' in string:
        string = removeSubStringAndSpace(string, 'FWD')
        return ('FWD', string)
    
    else: 
        return ('OTHER', string)

# remove cylinders (4C or 6C)    
# return the number of cylinders (if it exists in the string) and remove it from the string
def getNumberCylinders(string):
    if '4C' in string:
        string = removeSubStringAndSpace(string, '4C')
        return ('4C', string)    
    elif '6C' in string:
        string = removeSubStringAndSpace(string, '6C')
        return ('6C', string)
    
    else: 
        return ('OTHER', string)
    
# return the number of litres (if it exists in the string), 'None' otherwise
def getLitresSubstring(string):
    # return any substring that matches a digit followed by a . followed by nothing or digits or an L
    m = re.search('\d\.\d*L*', string)
    if m:
        return m.group(0)
    
    return 'None'

# remove the litres from a string
def removeLitresFromString(string):
    # get if they are litres are in the string
    litres = getLitresSubstring(string)
    if 'None' in litres:
        return string
    
    # else remove the litres from the string
    return removeSubStringAndSpace(string, litres)  
   
# split the Model column in a table
def splitModelColumn(table):
    # create 3 new arrays for the values to add
    engineSizes = []
    wheelDrives  = []
    cylinders = []
    models = []
    
    # for each row in the models
    for i in range(0, len(table.Model)):
        # get the next model name
        s = table.Model[i]
        
        # strip the engine size
        s = getEngineSize(s)
        engineSizes.append(s[0])
       
        # strip wheel drive
        s = getWheelDrive(s[1])
        wheelDrives.append(s[0])
        
        # strip number of cylinders
        s = getNumberCylinders(s[1])
        cylinders.append(s[0])
        
        # remove any litres
        s = removeLitresFromString(s[1])
        
        # append the reduced model name to the models name
        models.append(s)
        
    # set the new model names
    table['Model'] = models
    
    # add the new columns to the table
    table['EngineSize'] = engineSizes
    table['WheelDrives'] = wheelDrives
    table['Cylinders'] = cylinders
    
    # return the new table with the split Model column
    return table

# apply the model column change to both the auto and autotest tables
auto = splitModelColumn(auto)
autotest = splitModelColumn(autotest)
```

```python
# Split SUB-MODEL into Number of doors and Litres

# return the number of doors 
def getNumberDoors(string):
    if string in ['2D', '3D']:
        return '2D'
    elif string in ['4D', '5D']:
        return '4D'
    
    else: 
        return 'OTHER'
    
# split the Sub-Model column in a table
def splitSubModelColumn(table):
    # create new arrays for the values to add
    numberDoors = []
    litres  = []
    
    # for each row in the sub-models
    for i in range(0, len(table.SubModel)):
        # get the next sub model name
        submodel = table.SubModel[i]
        
        # if its a float, don't bother trying to cast it
        if isinstance(submodel, float):
            s = ""
        else:
            s = submodel
        
        # get the number doors and append to doors list
        doors = getNumberDoors(s)
        numberDoors.append(doors)
       
        # get the number of litres and append to litres list
        lits = getLitresSubstring(s)     
        litres.append(lits)
    
    # add the new columns to the table
    table['NumDoors'] = numberDoors
    table['Litres'] = litres
    
    # remove the Sub-Model column
    table = table.drop('SubModel', 1)
    
    # return the new table with the split Sub-Model column
    return table

# split the sub-model columns in the auto and autotest tables
auto = splitSubModelColumn(auto)
autotest = splitSubModelColumn(autotest)
    
```

```python
# make a list of numbers from 1 to length of data
nums = list(xrange(len(data)))

# set the random seed
random.seed(123)

# shuffle the list in-place "randomly"
random.shuffle(nums)

# get length of data
dataLen = len(data)

# split data into n parts
def chunkIt(seq, n): 
    avg = len(seq) / float(n) 
    out = [] 
    last = 0.0 
    
    while last < len(seq): 
        out.append(seq[int(last):int(last + avg)])
        last += avg 
        
    return out

# split the indexes into 10 chunks
chunks = chunkIt(nums, 10) 

# get the indices from all second dimension arrays, apart from the n-th array
def getOthers(values, nth):
    res = []
    
    # for each array
    for i in range(0, len(values)):
        # if not the n-th array
        if i != nth:
            # add all the values to the result array
            for x in range(0, len(values[i])):
                res.append(values[i][x])

    return res

# calculate residual sum sqaure
def rss(actual, predicts):
    result = 0
    tot = 0
    
    # get the difference squared of each pair in the lists
    for i in range(0, len(actual)):
        tot = actual[i] - predicts[i]
        tot = tot*tot
        result = result + tot
        
    # divide the result by the length
    result = result / len(actual)
    return result   

# convert column from data frame to ACTUAL array
def columnToArray(column):
    ca = []
    for i in range(0, len(column)):
        ca.append(column[i])
    return ca   

# convert column from data frame to ACTUAL array
def columnToArrayWithIndices(column, indices):
    ca = []
    for i in range(0, len(indices)):
        ca.append(column[indices[i]])
    return ca   

# run a kfold validation on a tree
def kfoldValidation(classTree):

    # list of evaluation results
    evalResults = []
    
    # for each 10 main parts (K)
    for i in range(0, 10): 
        # get copy of data
        trainingCV = data.copy()
        
        # get the next K (i-th) rows
        testCV = trainingCV.ix[chunks[i]]  
        
        # get the test actual IsBadBuy results
        actualResults = auto.IsBadBuy.ix[chunks[i]]
         
        # training = get the other 9k parts in one data frame
        trainingCV = trainingCV.ix[getOthers(chunks, i)]
        
        # target = IsBadBuy for the 9k parts  
        targetsCV = auto.IsBadBuy.ix[getOthers(chunks, i)]
        
        # get the fitted predictions from a tree
        predictsCV = fitAndPredict(classTree, trainingCV, targetsCV, testCV, "")
        
        # get the predictions is bad buy
        finalResults = predictsCV['IsBadBuy'][0:len(testCV)]
        
        # convert the final results and actual results DataFrames to actual arrays
        finalResultsArray = columnToArray(finalResults)
        actualResultsArray = columnToArrayWithIndices(actualResults, chunks[i])
        
        # run evaluation function on original targets and predicted targets
        tempRes = rss(finalResultsArray, actualResultsArray)
         
        # store value in results array
        evalResults.append(tempRes)
                
    return np.mean(evalResults)

# result pairs
resultPairs = {}

# get the k-fold score for the trees
resultPairs["Decision"] = kfoldValidation(tree.DecisionTreeClassifier())
resultPairs["RandomForest"] = kfoldValidation(RandomForestClassifier(n_estimators=3, min_samples_split=1, max_depth=None, random_state=0))
resultPairs["Extras"] = kfoldValidation(ExtraTreesClassifier(n_estimators=3, min_samples_split=1, max_depth=None, random_state=0))
resultPairs["AdaBoost"] = kfoldValidation(AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=10), algorithm="SAMME",n_estimators=20))
resultPairs["Gradient"] = kfoldValidation(GradientBoostingClassifier(max_depth=1, learning_rate=1.0, n_estimators=20, random_state = 0))

print resultPairs
```
